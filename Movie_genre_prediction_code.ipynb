```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
sep= ':::'
```

```python
train_data= pd.read_csv(r'C:\Users\lenovo\OneDrive\Desktop\Movie genre prediction\train_data.txt',sep=':::', names=['Title', 'Genre', 'Description'], engine='python')
test_data= pd.read_csv(r'C:\Users\lenovo\OneDrive\Desktop\Movie genre prediction\test_data.txt',sep=':::', names=['Title', 'Genre', 'Description'], engine='python')
test_data_solution= pd.read_csv(r'C:\Users\lenovo\OneDrive\Desktop\Movie genre prediction\test_data_solution.txt',sep=':::', names=['Title', 'Genre', 'Description'], engine='python')

```

```python
print("Train Data Columns and Samples")
print(train_data.columns)
print(train_data.head())

print("\nTest Data Columns and Sample:")
print(test_data.columns)
print(train_data.head())

print("\nTest data solution Columns and Sample:")
print(test_data_solution.columns)
print(test_data_solution.head())
```

```python
plot_column= 'Description'
genre_column='Genre'

train_data[plot_column].fillna('',inplace=True)
test_data[plot_column].fillna('',inplace=True)
```

```python
label_encoder= LabelEncoder()
train_data['genre_encoded']= label_encoder.fit_transform(train_data[genre_column])
```

```python
#Text vectorization using TF-IDF

tfidf_vectorizer= TfidfVectorizer(max_features=5000, stop_words='english')
X_train_tfidf= tfidf_vectorizer.fit_transform(train_data[plot_column])
X_test_tfidf= tfidf_vectorizer.transform(test_data[plot_column])
```

```python
#features and target variable
X= X_train_tfidf
Y= train_data['genre_encoded']
```

```python
# Split the data into training and validation data
X_train, X_val, Y_train, Y_val= train_test_split(X,Y, test_size=0.2, random_state=42)
```

```python
#Importing the Logistic Regression Model
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
```

```python
#Initialize and train the model
model= LogisticRegression(max_iter=1000)
model.fit(X_train, Y_train)
```

LOGISTIC REGRESSION


ACCURACY:


```python
#predict on the validation set

Y_val_predict= model.predict(X_val)

#Evaluate the model
print("Accuracy:", accuracy_score(Y_val,Y_val_predict))

```

```python
print(classification_report(Y_val, Y_val_predict, target_names= label_encoder.classes_))
```
